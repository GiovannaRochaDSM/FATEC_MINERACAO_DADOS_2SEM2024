# 01 - Criação e Manipulação de Arrays

Crie um array Numpy com 10 números inteiros aleatórios entre 1 e 20 e extraia os elementos pares.

```python
import numpy as np

# Criação do array
arrayAleatorio = np.random.randint(1, 21, size=10)
print("Array aleatório:", arrayAleatorio)

# Elementos pares do array
pares = arrayAleatorio[arrayAleatorio % 2 == 0]
print("Elementos pares do array:", pares)
```

# 02 - Operações Aritméticas com Arrays

Crie dois arrays de 10 números inteiros aleatórios entre 0 e 50 e realize operações de soma, subtração, multiplicação e divisão elemento a elemento.

```python
# Criação dos arrays
arrayUm = np.random.randint(51, size=10)
print("Array um:", arrayUm, "\n")

arrayDois = np.random.randint(51, size=10)
print("Array dois:", arrayDois, "\n")

# Soma
somaArrays = arrayUm + arrayDois
print("Soma dos arrays:", somaArrays, "\n")

# Subtração
subtracaoArrays = arrayUm - arrayDois
print("Subtração dos arrays:", subtracaoArrays, "\n")

# Multiplicação
multiplicacaoArrays = arrayUm * arrayDois
print("Multiplicação dos arrays:", multiplicacaoArrays, "\n")

# Divisão
divisaoArrays = arrayUm / arrayDois
print("Divisão dos arrays:", divisaoArrays, "\n")
```

# 03 - Indexação e Slicing

Dado um array Numpy com 20 números aleatórios entre 10 e 100, extraia os elementos entre a 5ª e a 15ª posição.

```python
# Criação dos arrays
array = np.random.randint(10, 101, size=20)
print("Array:", array, "\n")

# Extração dos elementos
elementosExtraidos = array[4:15]
print("Elementos extraídos:", elementosExtraidos)
```

# 04 - Manipulação de Matrizes

Crie uma matriz 3x3 com valores aleatórios entre 0 e 1, e encontre o valor máximo e mínimo.

```python
# Criação da matriz - A função rand já gera números aleatórios no intervalo 0 e 1
matriz = np.random.rand(3, 3)
print("Matriz:\n", matriz, "\n")

# Valor máximo
valorMaximo = np.max(matriz)
print("Valor máximo da matriz:", valorMaximo)

# Valor mínimo
valorMinimo = np.min(matriz)
print("Valor mínimo da matriz:", valorMinimo)
```

# 06 - Criação de DataFrames

Crie um DataFrame com dados de 5 pessoas, onde cada pessoa tem nome, idade (entre 20 e 40), e cidade (escolha 3 cidades aleatórias).

# 07 - Criação de DataFrame com Dados Sintéticos

Crie um DataFrame fictício de vendas contendo colunas como Produto, Quantidade (entre 1 e 20), e Preço (entre 10 e 100).

```python
# Dados
dadosVendas = {
    'Produto': ['Arroz', 'Feijão', 'Óleo', 'Açúcar', 'Sal', 'Macarrão', 'Leite', 'Café', 'Manteiga', 'Farinha'],
    'Quantidade': np.random.randint(1, 21, size=10),
    'Preço': np.random.uniform(10, 100, size=10).round(2)
}

# Criação do DataFrame
dfVendas = pd.DataFrame(dadosVendas)
dfVendas
```

# 08 - Filtragem de Dados

Filtre o DataFrame criado no exercício anterior para mostrar apenas as vendas onde a quantidade é maior que 10.

```python
# Filtro
dfVendasFiltrado = dfVendas[dfVendas['Quantidade'] > 10]
dfVendasFiltrado
```

# 09 - Operações em Colunas

Adicione uma coluna ao DataFrame com o valor total da venda (Quantidade * Preço).

```python
# Criação da coluna
total = dfVendas['Quantidade'] * dfVendas['Preço']

# Adição da coluna ao dfVendas
dfVendas = pd.concat([dfVendas, total.rename("Total")], axis=1)
dfVendas
```

# 10 - Agrupamento e Agregação

Agrupe o DataFrame pelo campo Produto e calcule a soma das quantidades vendidas para cada produto.

```python
# Agrupamento
dfAgrupado = dfVendas.groupby('Produto')['Quantidade'].sum()
dfAgrupado
```

# 11 - Gráfico de Linhas

Crie um gráfico de linha que mostra a variação de preços de um produto fictício ao longo de 12 meses.

```python
import matplotlib.pyplot as plt

meses = np.arange(1, 13)
precos = np.random.uniform(10, 100, size=12).round(2)

plt.plot(meses, precos, marker='o')
plt.title('Variação de Preços em 12 Meses')
plt.xlabel('Meses')
plt.ylabel('Preço')
plt.grid(True)
plt.show()
```

# 14 - Gráfico de Dispersão

Crie um gráfico de dispersão que mostre a relação entre altura e peso em um conjunto de dados fictício.

```python
altura = np.random.uniform(150, 190, size=100)
peso = np.random.uniform(50, 100, size=100)

plt.scatter(altura, peso, alpha=0.7)
plt.title('Relação entre Altura e Peso')
plt.xlabel('Altura (cm)')
plt.ylabel('Peso (kg)')
plt.show()
```

# 16 - Gráfico de Barras com Seaborn

Use Seaborn para criar um gráfico de barras com os mesmos dados fictícios de quantidade vendida por produto.

```python
import seaborn as sns

df = pd.DataFrame({
    'Produto': produtos,
    'Quantidade': quantidades
})

plt.figure(figsize=(10, 6))
grafico_barras = sns.barplot(x='Produto', y='Quantidade', data=df, palette='viridis')
grafico_barras.set_title('Quantidade Vendida por Produto')
grafico_barras.set_ylabel('Quantidade')
grafico_barras.set_xlabel('Produto')
grafico_barras.legend(['Quantidade'], loc='upper right')
plt.show()
```

# 17 - Gráfico de Dispersão com Seaborn

Recrie o gráfico de dispersão da relação entre altura e peso usando Seaborn, incluindo uma linha de regressão.

```python
dfAlturaPeso = pd.DataFrame({'Altura': altura, 'Peso': peso})

sns.regplot(x = 'Altura', y = 'Peso', data = dfAlturaPeso, scatter_kws = {'alpha':0.7}, line_kws = {'color':'yellow'})
plt.title('Relação entre Altura e Peso com Linha de Regressão')
plt.xlabel('Altura (cm)')
plt.ylabel('Peso (kg)')
plt.show()
```

# 19 - Heatmap

Crie um heatmap para mostrar a correlação entre diferentes variáveis em um DataFrame fictício.

```python
dfCorrelacao = pd.DataFrame({
    'A': np.random.rand(100),
    'B': np.random.rand(100),
    'C': np.random.rand(100),
    'D': np.random.rand(100)
})

matrizCorrelacao = dfCorrelacao.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(matrizCorrelacao, annot=True, cmap='coolwarm')
plt.title('Heatmap de Correlação')
plt.show()
```

# 20 - Pairplot

Crie um pairplot para explorar relações entre várias variáveis em um conjunto de dados fictício.

```python
dfCorr = pd.DataFrame({
    'A': np.random.normal(loc=0, scale=1, size=100), # loc (Média), scale (desvio padrão)
    'B': np.random.normal(loc=1, scale=2, size=100),
    'C': np.random.normal(loc=-1, scale=0.5, size=100),
    'D': np.random.normal(loc=2, scale=1.5, size=100)
})


sns.pairplot(dfCorr)
plt.suptitle('Pairplot de Variáveis')
plt.show()
```
